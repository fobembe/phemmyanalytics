```python
data[["sector", "Large_Firm","Female_BM","Mgrs_educ","Wkr_Exp_Mgr","Geo_Zone","Mgr_Shareholding","Foreign_BM"]] = df[["sector", "Large_Firm",
                                                                                                                      "Female_BM","Mgrs_educ","Wkr_Exp_Mgr",
                                                                                                                      "Geo_Zone","Mgr_Shareholding","Foreign_BM"]].astype("category")
```

```python
%%stata -d data
logit Product i.sector Large_Firm Female_BM Mgrs_educ Wkr_Exp_Mgr Geo_Zone Mgr_Shareholding Foreign_BM
```

```python
%%stata -d data
logit Product i.sector i.Large_Firm i.Female_BM i.Mgrs_educ i.Wkr_Exp_Mgr i.Geo_Zone i.Mgr_Shareholding i.Foreign_BM
estimates store Product_REG
logit Process i.sector i.Large_Firm i.Female_BM i.Mgrs_educ i.Wkr_Exp_Mgr i.Geo_Zone i.Mgr_Shareholding i.Foreign_BM
estimates store Process_REG

esttab Product_REG Process_REG, b(%10.4f) se scalars(N r2 F ll) mtitles
```

```python
%%stata -d data
eststo: quietly logit Product i.sector i.Large_Firm i.Female_BM i.Mgrs_educ i.Wkr_Exp_Mgr i.Geo_Zone i.Mgr_Shareholding i.Foreign_BM
estimates store Product_REG
eststo: quietly logit Process i.sector i.Large_Firm i.Female_BM i.Mgrs_educ i.Wkr_Exp_Mgr i.Geo_Zone i.Mgr_Shareholding i.Foreign_BM
estimates store Process_REG

esttab using regoutput2.rtf
```
